```python
def large_sum(s1,s2):
    max_len = max(len(s1),len(s2))
    sum_arr = [0]*(max_len+1)
    i = max_len+1
    end_s1, end_s2 = len(s1)-1, len(s2)-1
    carry = 0
    while(1):
        summ = 0
        if end_s1 >= 0 and end_s2 >= 0:
            summ = int(s1[end_s1]) + int(s2[end_s2]) + carry
            if summ > 9:
                carry = summ/10
                summ %= 10
            sum_arr[i] = summ
            i -= 1
            end_s1 -= 1
            end_s2 -= 1

        elif end_s1 >= 0:
            summ = int(s1[end_s1])
            sum_arr[i] = summ

        elif end_s2 >= 0:
            summ = int(s2[end_s2])
            sum_arr[i] = summ
        
        else:
            break
    
    cnt = 0
    for j in range(len(sum_arr)):
        if sum_arr[j] == 0:
            cnt += 1
        else:
            break

    res = "".join(sum_arr)
    return res

```
